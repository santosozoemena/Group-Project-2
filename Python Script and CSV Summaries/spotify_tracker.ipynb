```python
# Dependencies
import csv
import pandas as pd
import os
import numpy as np
from datetime import datetime
```

```python
# Read csv
csvpath = os.path.join('Desktop', 'Project-2', 'spotify_data.csv')
import csv

spotify_df = pd.read_csv(csvpath, encoding="ISO-8859-1")
```

```python
# Record starting date and ending date of data in question
```

```python
# Add a normalizing column for each song in each region to convert to streams per million of populace
spotify_df["Streams per 1 million"] = (spotify_df["Streams"]/spotify_df["Population"]*1000000).round(2)
```

```python
# Drop the obsolete "Stream % of Pop" column
del spotify_df['Stream % of Pop']
```

```python
# Add a blank column to flag number one hits
spotify_df['Number One']=np.nan
```

```python
spotify_df['Date'] = pd.to_datetime(spotify_df['Date'])
spotify_df['RELEASE'] = pd.to_datetime(spotify_df['RELEASE'])
```

```python
spotify_df.head()
```

```python
spotify_df.count()
```

```python
# Create a new dataframe that retains only songs that are in the number one position (for each region)
number_one_df = spotify_df.loc[spotify_df["Position"] == 1,:]
```

```python
number_one_df.head(3)
```

```python
number_one_charts_df = number_one_df.groupby(["Region", "Track Name", "Artist"]).size().reset_index(name='Days at # 1')
number_one_charts_df
```

```python
# SAVE A CSV OF THE TABLE ("number_one_chart.csv") TO COMBINE WITH "song_tracker.csv" LATER
number_one_charts_df.to_csv("Desktop/Project-2/number_one_chart.csv", index=False, header=True)
```

```python
# Drop all dupclicates so only one instance of each song remains
hits_df = number_one_df.drop_duplicates(subset=['URL','Region'], keep ='first')
```

```python
hits_df.head(3)
```

```python
# Create a new stripped-down dataframe for merging later.
hit_filter_df = hits_df[['URL', 'Region', 'Number One']]
```

```python
hit_filter_df.head()
```

```python
hit_filter_df.count()
```

```python
# Export file as a csv, without the Pandas index, but with the header
hit_filter_df.to_csv("Desktop/Project-2/numberOneUnique.csv", index=False, header=True)
```

```python
# Read the newly created csv back in as "numberOneUnique_df" for later merging with the original "spotify_df"
csvpath2 = os.path.join('Desktop', 'Project-2', 'numberOneUnique.csv')
import csv

numberOneUnique_df = pd.read_csv(csvpath2, encoding="ISO-8859-1")
```

```python
numberOneUnique_df.head(3)
```

```python
# Fill the "Number One" Column in the "numberOneUnique_df" with "Number One Hit" 
numberOneUnique_df['Number One'] = 'Number One Hit'
```

```python
# merge the the newly imported "numberOneUnique_df" with the original "spotify_df" df one to id the songs that make it to # 1
merge_table_df = pd.merge(spotify_df, numberOneUnique_df, on=['URL','Region'], how='left')
```

```python
merge_table_df
```

```python
# Clean up the merge results
clean_merge_df = merge_table_df.drop(["Number One_x"], axis=1)
clean_merge_df.rename(columns={'Number One_y': 'Reached Number One'}, inplace=True)
```

```python
clean_merge_df.head(3)
```

```python
# SAVE A CSV OF THE NEW SPOTIFY TABLE ("spotify_merged.csv") WHICH NOW FLAGS SONGS THAT REACH NUMBER ONE IN A REGION
clean_merge_df.to_csv("Desktop/Project-2/spotify_merged_data.csv", index=False, header=True)
```

```python
# Create a dataframe that just consists of the songs that make it to number one in at least one region
hit_tracker_df = clean_merge_df.loc[clean_merge_df["Reached Number One"] == "Number One Hit",:]
```

```python
hit_tracker_df.head(3)
```

```python
# SAVE A CSV WHICH NOW ONLY SHOWS DATA FOR SONGS THAT REACHED NUMBER ONE
hit_tracker_df.to_csv("Desktop/Project-2/hit_songs_only.csv", index=False, header=True)
```

```python
hit_tracker_grouped_df = hit_tracker_df.groupby(["Region"]).agg({'Date': [min, max],
                                                                'Streams': [max],
                                                                'Streams per 1 million': [max]})
hit_tracker_grouped_df.columns = ["_".join(x) for x in hit_tracker_grouped_df.columns.ravel()]
```

```python
hit_tracker_grouped_df["Days Charted"] = (hit_tracker_grouped_df["Date_max"] - hit_tracker_grouped_df["Date_min"])
hit_tracker_grouped_df = hit_tracker_grouped_df.reset_index()
hit_tracker_grouped_df
```

```python
# SAVE A CSV OF THE NEW SPOTIFY TABLE ("hit_tracker.csv") WHICH NOW FLAGS SONGS THAT REACH NUMBER ONE IN A REGION
hit_tracker_grouped_df.to_csv("Desktop/Project-2/hit_tracker.csv", index=False, header=True)
```

```python
song_tracker_grouped_df = hit_tracker_df.groupby(["Region", "Artist", "Track Name"]).agg({'Date': [min, max],
                                                                                         'Streams': [max],
                                                                                         'Streams per 1 million': [max]})
song_tracker_grouped_df.columns = ["_".join(x) for x in song_tracker_grouped_df.columns.ravel()]
```

```python
song_tracker_grouped_df["Days Charted"] = (song_tracker_grouped_df["Date_max"] - song_tracker_grouped_df["Date_min"])
song_tracker_grouped_df = song_tracker_grouped_df.reset_index()
song_tracker_grouped_df
```

```python
# SAVE A CSV OF THE NEW SPOTIFY TABLE ("song_tracker.csv") WHICH NOW FLAGS SONGS THAT REACH NUMBER ONE IN A REGION
song_tracker_grouped_df.to_csv("Desktop/Project-2/song_tracker.csv", index=False, header=True)
```
